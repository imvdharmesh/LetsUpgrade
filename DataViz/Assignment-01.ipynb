## Data Visualization Using Python

```python
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn, randint, uniform, sample

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Question 1 : Plot a line plot between a and b:

```python
a=np.arange(40,50)
b=np.arange(50,60)

plt.plot(a,b,'bo-')
plt.title("Line Plot")
plt.xlabel('a')
plt.ylabel("b")
```

## Question 2 : Plot a line plot showing the sales trend in company 1 and 2:

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2

plt.plot(days, sales_1, 'bo--', label='Company 1')
plt.plot(days, sales_2, 'rv--', label='Company 2')
plt.title('Sales Trend')
plt.xlabel("Days of Week")
plt.ylabel("Sales of Company")
plt.legend()
```
