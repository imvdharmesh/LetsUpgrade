### Question 1 : Plot using sns.relplot, line plot, box plot on fmri data set.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
fmri.head(5)
```

## relplot

```python
sns.relplot(x = 'timepoint', y = 'signal', data = fmri, hue='event')
```

## lineplot

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, sort = True)
```

## boxplot

```python
sns.boxplot(x = 'region', y = 'signal', data = fmri, hue = 'event')
```
